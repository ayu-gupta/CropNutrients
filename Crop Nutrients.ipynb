```python
pro=pd.read_excel('Desktop/crop_production.csv.xlsx')
pro
```

```python
States=pro.State_Name.unique()
Season=pro.Season.unique()
Crop=pro.Crop.unique()
Season
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
le.fit(States)
S_transform=le.transform(States)
S_transform
```

```python
for i in range(0,33):
    pro.replace(to_replace=States[i], value=S_transform[i],inplace= True )
```

```python
from sklearn import preprocessing
les=preprocessing.LabelEncoder()
les.fit(Season)
Se_transform=les.transform(Season)
Se_transform
```

```python
for i in range(0,6):
    pro.replace(to_replace=Season[i], value=Se_transform[i],inplace= True )
```

```python
from sklearn import preprocessing
lec=preprocessing.LabelEncoder()
lec.fit(Crop)
C_transform=lec.transform(Crop)
C_transform.size
```

```python
for i in range(0,124):
    pro.replace(to_replace=Crop[i], value=C_transform[i],inplace= True )
```
